```python
from bs4 import BeautifulSoup
import json, io

f=open("publist_super_clean.html", 'r')
soup= BeautifulSoup(f.read(),'lxml')
```

```python
det=soup.find_all('li')
```

```python
data1=[]
data2=[]
data3=[]
data4=[]
data5=[]

for z in range(len(det)):
    url=(det[z].find('a'))
    if url is not None:
        url=url.attrs['href']
        data1.append(url)
        print(url)
    else:
        data1.append('-')
        
    
    name=(det[z].find('a'))    
    if name is not None:
        name=name.text
        name=name.replace('\n','')
        data2.append(name)
        print(name)
    else:
        data2.append('-')
    
    
    
    author0=(det[z].find('a'))
    if author0 is not None:
        author=(((((author0)).next_element).next_element).next_element).next_element
        author=author.replace('\n','')
        data3.append(author)
        print (author)
    else:
        data3.append('-')
    
    journal0=(det[z].find('i'))
    
    if journal0 is not None:
        journal=journal0.text
        journal=journal.replace('\n','')
        data4.append(journal)
        print(journal)

    else:
        data4.append('-')
    

    volume=(det[z].find('b'))
    if volume is not None:
        volume=volume.text
        data5.append(volume)
        print(volume)
    else:
        data5.append('-')
    print("++++++++++++++++++++++++")
```

```python
jsonList = []
for i in range(0,len(data1)):
    jsonList.append({'Url': data1[i],'Name': data2[i],'Authors': data3[i],'Journal': data4[i],'Volume': data5[i]})
```

```python
jsonList
```

```python
with io.open('result_output.txt', 'w', encoding='utf8') as json_file:
    data3 = json.dumps(jsonList, ensure_ascii=False, encoding='utf8',indent=4, sort_keys=True)
    json_file.write(unicode(data3))
```
